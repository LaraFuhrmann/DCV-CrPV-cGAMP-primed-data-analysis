```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dcv_homopolymeric_region_rdrp.csv")
```

```python
df['del_in_region'] = df['del_in_region'].astype(int)
df['insert_in_region'] = df['insert_in_region'].astype(int)
df['insert_del_in_region'] = df['insert_in_region'] * df['del_in_region']

df['del_region_close_to_read_end'] = df['del_region_close_to_read_end'].astype(int)

#df['no_del_in_region'] = 1 - df['del_in_region']
#df['qual_homo_region_with_del'] = df['qual_del_in_region']*df['del_in_region']
#df['qual_homo_region_no_del'] = df['qual_del_in_region']*(1-df['del_in_region'])
#df['homo_close_read_edge_with_del'] = df['del_region_close_to_read_end']*df['del_in_region']
#df['homo_close_read_edge_no_del'] = df['del_region_close_to_read_end']*(1-df['del_in_region'])

df['coverage'] = 1
```

```python
df_qual_del = pd.pivot_table(data= df, 
                             values=['del_in_region', 
                                     'insert_in_region', 
                                     'insert_del_in_region', 
                                     #'no_del_in_region',
                                     'qual_del_in_region', 
                                     'del_region_close_to_read_end', 
                                     #'qual_homo_region_with_del', 
                                     #'qual_homo_region_no_del',
                                     #'homo_close_read_edge_with_del',
                                     'coverage',
                                     #'homo_close_read_edge_no_del'
                                    ], 
                             index=['file'], 
                             aggfunc={'del_in_region': np.mean, 
                                      'insert_in_region': np.mean, 
                                      'insert_del_in_region': np.mean, 
                                      #'no_del_in_region': np.sum,
                                      'del_region_close_to_read_end': np.mean, 
                                      'qual_del_in_region': np.mean,
                                      'coverage': np.sum, 
                                      #'qual_homo_region_with_del': np.sum, 
                                      #'qual_homo_region_no_del': np.sum,
                                      #'homo_close_read_edge_with_del': np.sum,
                                      #'homo_close_read_edge_no_del': np.sum,
                             }
                            )
df_qual_del = df_qual_del.reset_index()
```
